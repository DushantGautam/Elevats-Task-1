```python
# Define mapping for standardization
marital_map = {
    'Single': 'Single',
    'Alone': 'Single',
    'YOLO': 'Single',
    'Absurd': 'Single',

    'Married': 'Married',
    'Together': 'Married',

    'Divorced': 'Divorced',
    'Widow': 'Widowed'
}
```

```python
marital_map
```

```python
df['Marital_Status'] = df['Marital_Status'].map(marital_map)
```

```python
df['Marital_Status'].unique()
```

```python
df.columns = df.columns.str.lower()
```
